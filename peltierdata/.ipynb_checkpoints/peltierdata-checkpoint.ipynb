```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


coldntc=[] #in celsius
hotntc=[] #in celsius
digitaltemp=[] #in celsius
time=[] #in milliseconds


coldntc, hotntc, time = np.loadtxt('pid-5c.dat', unpack=True)

time1=[x/1000. for x in time]
setpoint=[-5] * (len(time))

err=[x-y for x,y in zip(coldntc,setpoint)]
deltat=[abs(x-y) for x,y in zip(coldntc,hotntc)]

plt.figure(figsize=(15,10))
plt.rcParams.update({'font.size': 18})
plt.plot(time1, coldntc, label='Cold NTC Data', color='b', linewidth=3)
plt.plot(time1, hotntc, label='Hot NTC Data', color='r', linewidth=3)
plt.plot(time1, deltat, label='$\Delta$T', color='g', linewidth=3)
plt.plot(time1, setpoint, label='Setpoint = -5$\degree$C', color='C1', linewidth=3)
#plt.xlim(200,600)
#plt.ylim(-.25,.2)
plt.ylabel('Temperature ($\degree$C)')
plt.xlabel('Time (seconds)')
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1), ncol=4)
plt.title('No Heat Load with Setpoint = -5$\degree$C')
plt.show()

```
